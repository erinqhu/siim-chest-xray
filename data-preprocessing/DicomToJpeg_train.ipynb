```
inputdir = "/content/gdrive/My Drive/siim-acr/dicom-images-train"
outdir = "train"
```

```
# get all dcm files from the dicom-images-train folder
def getAllFiles(path, type = '.dcm'):
    dcmFiles = []
    for dirpath, subdirs, files in os.walk(path):
        for x in files:
            if x.endswith(type):
                dcmFiles.append([dirpath, x])
    return dcmFiles

allDCMImages = getAllFiles(inputdir)
allDCMImages
```

```
for file in allDCMImages:
    filename = os.path.join(file[0],file[1])
    ds = pydicom.read_file(filename)  # read dicom image
    img = ds.pixel_array  # get image array

    #print(filename)
    if not os.path.exists(outdir):
        os.makedirs(outdir)

    cv2.imwrite(outdir + "/" + file[1].replace('.dcm', '.jpg'), img)  # write jpg image
```
