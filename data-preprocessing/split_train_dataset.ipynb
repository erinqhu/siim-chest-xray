```
# get all file names
def getAllFiles(path, type = '.dcm'):
    dcmFiles = []
    for dirpath, subdirs, files in os.walk(path):
        for x in files:
            if x.endswith(type):
                dcmFiles.append(x[:-4])
    return dcmFiles
```
